# Box plot : 

A box plot (or box-and-whisker plot) s is the visual representation of the depicting groups of numerical data through their quartiles against continuous/categorical data.

A box plot consists of 5 things.

Minimum

First Quartile or 25%

Median (Second Quartile) or 50%

Third Quartile or 75%

Maximum

```python
sns.boxplot( data['Age'] )
```

```python
sns.boxplot( data['Age'], data['Weight'])
```

# Violion Plot

A violin plot is similar to a boxplot. It shows several quantitative data across one or more categorical variables such that those distributions can be compared.

```python
# Bar chart 
```

```python
x = [3, 2, 5, 12, 4, 5, 2]
y = [2, 3, 4, 1, 3, 5, 6]
```

```python
# this will plot a simple bar chart
plt.bar(x, y)

# Title to the plot
plt.title("Bar Chart")

#Adding legends
plt.legend(['bar'])
plt.show()
```

```python
# Histogram
```

```python
x = [1, 2, 3, 4, 5, 6, 7, 8, 4]
```

```python
# this will plot a simple histogram
plt.hist(x, bins = [1, 2, 3, 4, 5, 6, 7])

# title to the plot
plt.title("Histogram")

#adding the legends
plt.legend(['bar'])
plt.show
```

```python
# Scatter plot
```

```python
x = [3, 1, 3, 12, 2, 4, 4]
y = [3, 2, 1, 4, 5, 6, 7]
```

```python
# This will plot a simple scatter chart
plt.scatter(x, y)

# Adding legend to the plot
plt.legend("A")

# Title to the plot
plt.title("Scatter chart")
plt.show()
```

```python
# Stack plot
```

```python
days = [1, 2, 3, 4, 5]
studying = [7, 8, 6, 11, 7] # study hour
playing = [8, 5, 7, 8, 13] # play hour
```

```python
# stackplot with x, y , colors value
plt.stackplot(days, studying, playing, colors = ['r', 'c'])

plt.xlabel('Days')

plt.ylabel('No of Hours')

# title of graph
plt.title('Representation of study and \ playing wrt to Days')
plt.show()
```

```python
# Box plot
```

```python
np.random.seed(10)
data = np.random.normal(100, 20, 200)
fig = plt.figure(figsize = (10, 7))
plt.boxplot(data)
plt.show()
```
